```python
def train_Q_learner(num_trials = 1000, k=2, n=4, grid_size = 7):
    """
    Trains the Q Learner against Minimax Depth k
    
    Inputs:
    Number of games to play
    On grid_size x
    N tokens to connect
    
    Outputs:
    Q Learner value table after training
    """
    depth = k
    N = n
    grid_size = grid_size
    x = ConnectN(grid_size, N)
    
    p1 = Q_Learner(x, None, True, player=1)
    p2 = Random_Learner(x)
    play_game_no_output(x, p1, p2)
    
    for game in xrange(1, num_trials):
                
        x = ConnectN(grid_size, N)
        p1 = Q_Learner(x, p1.value_table, True, player=1)
        p2 = Random_Learner(x)
        play_game_no_output(x, p1, p2)

        if game == num_trials - 1:
            return p1.value_table
```

```python
full_hist = np.concatenate((Q_hist_Q, Q_hist_rand, Q_hist_MMd2, Q_hist_MMd4, Q_hist_rand2))
full_hist_filter = np.zeros(full_hist.shape[0])
for i in xrange(0,len(full_hist_filter)):
    if full_hist[i] == 1:
        full_hist_filter[i] = 1
    if full_hist[i] == 2:
        full_hist_filter[i] = -1

cumulative_history = np.cumsum(full_hist_filter)
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(0,10000),cumulative_history[0:10000],linewidth=3.0)
plt.plot(range(10000,20000),cumulative_history[10000:20000],linewidth=3.0)
plt.plot(range(20000,21000),cumulative_history[20000:21000],linewidth=3.0)
plt.plot(range(21000,22000),cumulative_history[21000:22000],linewidth=3.0)
plt.plot(range(22000,32000),cumulative_history[22000:32000],linewidth=3.0)
plt.legend(['Q-Learning (w/ known states) vs. Q-Learning', 'Q vs. Random', 'Q vs. Minimax Depth-2','Q vs. Minimax Depth-4', 'Q vs. Random'], loc=0)
plt.title('Q-Learning (w/ known states) Training Period Cumulative Rewards')
plt.xlabel('Games')
plt.ylabel('Cumulative Rewards (1 for win and -1 for loss)')
plt.savefig('knownQ_training.png', dpi=500)

#print full_hist_filter[-100:]

#Q_hist_Q
#Q_hist_rand
#Q_hist_MMd2
#Q_hist_MMd4
#Q_hist_rand2
```
